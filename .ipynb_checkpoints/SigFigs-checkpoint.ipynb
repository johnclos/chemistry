```python
# create the dataframe and variable for the ouput
SigFigs_output_df = pd.DataFrame()
SigFigs_output = []

# specify the number of significant figure
significant_digits = 3

for i, j in SigFigs_output_df.iterrows():
    val = (i,j)[1]
    SigFig_number =  round(val, significant_digits - int(math.floor(math.log10(abs(float(val))))) - 1)
    #print(SigFig_number)
    SigFigs_output.append(("%.3f" % SigFig_number[0]).rstrip('0').rstrip('.'))

SigFigs_output
SigFigs_output_df['Output'] = SigFigs_output
SigFigs_output_df
```

```python
SigFigs_output_df
SigFigs_output = []
significant_digits = 3
for i, j in SigFigs_df.iterrows():
    val = (i,j)[1]
    SigFig_number =  round(val, significant_digits - int(math.floor(math.log10(abs(float(val))))) - 1)
    #print(SigFig_number)
    SigFigs_output.append(("%.99f" % SigFig_number[0]).rstrip('0').rstrip('.'))

SigFigs_output_df['Output'] = SigFigs_output
SigFigs_output_df

for i, j in SigFigs_output_df.iterrows():
    if float(SigFigs_output_df['Output'][1]) > 0.0001:
        SigFigs_output_df['Output'][1] = ("%.6f" % float(SigFigs_output_df['Output'][1])).rstrip('0').rstrip('.')
    else:
        SigFigs_output_df['Output'][1] = ("%.10f" % float(SigFigs_output_df['Output'][1])).rstrip('0').rstrip('.')

SigFigs_output_df
```
